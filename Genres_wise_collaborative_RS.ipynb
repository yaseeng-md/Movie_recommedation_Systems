```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import ast
import os
delete_index = [19730,29503,35587] # Delete these indexes to not face any errors
# movie_data.rename(columns={"id":"movieId"},inplace=True)
```

```python
movie_data = pd.read_csv(r"C:\Users\gandl\Documents\Projects\Movie_recommendations\data\movies_metadata.csv")
movie_data.rename(columns={"id":"movieId"},inplace=True)
movie_data.drop(index=delete_index,inplace=True,axis=0)
movie_data["movieId"] = movie_data["movieId"].astype(int)
movie_data
```

```python
def dict_to_arr_genres(series):
    s = ast.literal_eval(series)
    a = []
    for i in range (len(s)):
        a.append(s[i]['name'])
    return a
```

```python
movie_data["genres"] = movie_data["genres"].apply(dict_to_arr_genres)
movie_data
```

```python
genres_dataset = movie_data[["movieId","genres"]]
genres_dataset
```

```python
genres_dataset.info()
```

```python
genres_dataset
```

```python
exploded_genres = genres_dataset.explode("genres")
exploded_genres
```

```python
pt = exploded_genres.pivot_table(index="movieId",columns="genres",aggfunc="size",fill_value=0)
pt
```

```python
exploded_genres[exploded_genres["id"] == 2]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```
